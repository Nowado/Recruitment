```python
df_2016['Date'].unique()

dates = []
for m in [12,1]:
    for d in range(31):
        dates.append(str(m)+'-'+str(d+1))
```

```python
import datetime

def comparative_chart(df1,df2,column,summ):
    labels = dates
    
    if summ:
        first = df1.groupby('Date')[column].sum().values
        second = df2.groupby('Date')[column].sum().values
    else:
        first = df1.groupby('Date')[column].mean().values
        second = df2.groupby('Date')[column].mean().values
    
    
    x = np.arange(len(labels)) 
    width = 0.35

    fig, ax = plt.subplots()
        
    
    plt.plot(labels,first)
    plt.plot(labels,second)
    plt.bar(labels,first-second, color='green')

    start, end = ax.get_xlim()
    ax.xaxis.set_ticks(np.arange(start, end, 7))
    
    fig.tight_layout()
    fig.suptitle(column+ ' for years 2016, 2017 and difference',
             size=14, verticalalignment='top', horizontalalignment = 'center',y=1.1)
    plt.show()
```

```python
for ver in [('Net Sales (USD)',True), ('Discount',False),('Net Units Sold',True)]:
    print(ver)
    comparative_chart(df_2016,df_2017, ver[0],ver[1])
```

```python
df_2016_sales = df_2016[733:2049]
df_2017_sales = df_2017.loc[4833:5992]
```

```python
dates_dict_2016 = {}
dates_dict_2017 = {}

i=0
for date in df_2016_sales['Date'].unique():
    i+=1
    dates_dict_2016[date] = i

i=0
for date in df_2017_sales['Date'].unique():
    i+=1
    dates_dict_2017[date] = i

df_2016_sales = df_2016_sales.replace({"Date": dates_dict_2016})
df_2017_sales = df_2017_sales.replace({"Date": dates_dict_2017})
```

```python
reg_date_2016=df_2016_sales.groupby(['Region','Date'])['Net Sales (USD)'].sum()
reg_date_2017=df_2017_sales.groupby(['Region','Date'])['Net Sales (USD)'].sum()
```

```python
reg_date_2016 = reg_date_2016.unstack(level=0).fillna(0)
reg_date_2016.loc[14] = 0
reg_date_2016.loc[15] = 0
reg_date_2016['Total'] = reg_date_2016.sum(axis=1)
reg_date_2017 = reg_date_2017.unstack(level=0).fillna(0)
reg_date_2017['Total'] = reg_date_2017.sum(axis=1)
reg_date_2016.shape, reg_date_2017.shape
```

```python
fig, axs = plt.subplots(3, 4, figsize=(6,6), sharex=False)

i=0
j=0
for region in reg_date_2016.columns:
    
    print(i)
    dates = reg_date_2016.index
    y_pos = np.arange(len(dates))+1
    sales2016 = reg_date_2016[region]/10000
    sales2017 = reg_date_2017[region]/10000
    
    axs[j,i].barh(y_pos, sales2017, color = 'red', height = .5)
    axs[j,i].barh(y_pos, sales2016, color = 'none', edgecolor = 'green', height = .5)
    
    axs[j,i].set_title(region, rotation=0, fontsize=12)
    axs[j,i].tick_params(
    axis='x',           # changes apply to the x-axis
    which='both',       # both major and minor ticks are affected
    bottom=True,
    top=False,
    labelbottom=True)
        
    i+=1
    if i==4:
        i=0
        j+=1



plt.ylim(2)
plt.subplots_adjust(left=0, right=1, top = 1.5, )
fig.suptitle('Comparative net sales ($10,000) over days of sale, divided by region',
             size=16, verticalalignment='top', horizontalalignment = 'center', y=1.7)

plt.legend(handles=(), labels = (2016,2017))
plt.show()
```
